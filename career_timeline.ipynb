```python
import plotly.express as px
import pandas as pd
from google.colab import auth
import gspread
from google.auth import default
import plotly.figure_factory as ff
```

```python
# Authenticate user for google sheets
auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)
```

```python
worksheet = gc.open('career_timeline').sheet1
```

```python
# get_all_values gives a list of rows.
rows = worksheet.get_all_values()
```

```python
# Convert to a DataFrame and render.
df=pd.DataFrame.from_records(data=rows[1:len(rows)], columns=rows[0])
```

```python
df.head()
```

```python
df['full_title']=df['title'].astype(str) + ' -- ' + df['company'].astype(str)
```

```python
df.head()
```

```python
df_ff=df[['full_title', 'start_year', 'end_year', 'career_type']]
df_ff.columns=["Task", "Start", "Finish", "Career Type"]
```

```python
colors = {
    'Education': 'rgb(30, 203, 225)',
    'Professional': 'rgb(225, 30, 203)',
    'Independent':'rgb(203, 225, 30)'
          }
```

```python
fig=ff.create_gantt(df_ff, colors=colors, index_col="Career Type",
                    show_colorbar=True, title="Dr. C's Crazy Career Timeline",
                    height=700
)

fig.update_layout(
    title={
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'})
fig.update_layout(template='plotly_dark')
fig.update_xaxes(showgrid=True)
fig.update_yaxes(showgrid=True)
ticktext = [colors[x] for x in df_ff["Career Type"]]
fig.layout.xaxis.rangeselector = None
fig.show()
```
